```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
def logit(x):
    return 1/(1+np.exp(-np.array(x)))
```

```python

fig,ax=plt.subplots(3,3,figsize=(18,18))
c=0

for x0 in [-4,-3,-2,-0.5,0,0.5,2,3,4]:
    row=int(c/3)
    col=(c%3)
    v=abs(0.25*x0) if (x0 !=0) else 0.5
    nums=int(1e6)
    skew_scale= abs(v/x0) if x0!=0 else 0  
    ax[row][col].hist(logit(ss.norm.rvs(loc=x0,scale=v,size=nums)),alpha=0.5,label="normal",bins=100,density=True)
    ax[row][col].hist(logit(ss.skewnorm.rvs(-skew_scale*x0,loc=x0,scale=v,size=nums)),alpha=0.5,label="skew normal",bins=100,density=True)

    ax[row][col].axvline(logit(x0))
   # ax[row][col].legend(loc='best')
    print(logit(x0))
    c+=1
```

```python
X=np.linspace(-10,10)
plt.plot(X,logit(X))
```
