# Input parameters

```python
fig, ax=plt.subplots(figsize=(6,4))
rt.plot_histogram(df['s'],ax,"Salinity [g/L]",ymax=300)
ax.text(0.1,0.9,"(a)",transform=ax.transAxes)
plt.savefig("Full_System/Romania/Romania_Salinity_input.png",dpi=300,bbox_inches='tight')

fig, ax=plt.subplots(figsize=(6,4))
rt.plot_histogram(df['t'],ax,"Temperature [°C]",ymax=500)
ax.text(0.05,0.9,"(b)",transform=ax.transAxes)
plt.savefig("Full_System/Romania/Romania_Temperature_input.png",dpi=300,bbox_inches='tight')

fig, ax=plt.subplots(figsize=(6,4))
rt.plot_histogram(df['q'],ax,"Flow rate [L/s]",ymax=500)
ax.text(0.05,0.9,"(c)",transform=ax.transAxes)
plt.savefig("Full_System/Romania/Romania_Flowrate_input.png",dpi=300,bbox_inches='tight')

fig, ax=plt.subplots(figsize=(6,4))
rt.plot_histogram(df['ph'],ax,"pH [-]",ymax=300)
ax.text(0.1,0.9,"(d)",transform=ax.transAxes)
plt.savefig("Full_System/Romania/Romania_ph_input.png",dpi=300,bbox_inches='tight')
```
